```python
d1 = {}
d1['a'] = 'A'
d1['b'] = 'B'
d1['c'] = 'C'
d1['d'] = 'D'
d1['e'] = 'E'

d2 = {}
d2['a'] = 'A'
d2['c'] = 'C'
d2['e'] = 'E'
d2['b'] = 'B'
d2['d'] = 'D'

d3 = {}
d3['a'] = 'A'
d3['b'] = 'B'
d3['c'] = 'C'
d3['d'] = 'D'
d3['e'] = 'E'

print(d1, type(d1))
print("\n__________\n")

print(d1 == d2)
print(d1 == d3)
print("\n__________\n")

for k, v in d1.items():
    print("Key: |", k, "|, value: |", v, "|")
print("\n__________\n")

for k, v in d2.items():
    print("Key: |", k, "|, value: |", v, "|")
print("\n__________\n")

for k, v in d3.items():
    print("Key: |", k, "|, value: |", v, "|")
print("\n__________\n")
```

```python
from collections import OrderedDict

d1 = OrderedDict()
d1['a'] = 'A'
d1['b'] = 'B'
d1['c'] = 'C'
d1['d'] = 'D'
d1['e'] = 'E'

d2 = OrderedDict({})
d2['a'] = 'A'
d2['c'] = 'C'
d2['e'] = 'E'
d2['b'] = 'B'
d2['d'] = 'D'

d3 = OrderedDict()
d3['a'] = 'A'
d3['b'] = 'B'
d3['c'] = 'C'
d3['d'] = 'D'
d3['e'] = 'E'

print(d1, type(d1))
print("\n__________\n")

print(d1 == d2)
print(d1 == d3)
print("\n__________\n")

for k, v in d1.items():
    print("Key: |", k, "|, value: |", v, "|")
print("\n__________\n")

for k, v in d2.items():
    print("Key: |", k, "|, value: |", v, "|")
print("\n__________\n")

for k, v in d3.items():
    print("Key: |", k, "|, value: |", v, "|")
print("\n__________\n")
```
